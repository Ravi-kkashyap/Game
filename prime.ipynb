```python
def isprime(num):
    if num > 1:
    # check for factors
        for i in range(2,num):
            if (num % i) == 0:
                 return False
        return True
    else:
        return False
    
def get_primes(upto):
    prime_list = []
    num=1
    while num<=upto:
        if isprime(num):
            prime_list.append(num)
        num+=1
    return prime_list

def find_largest_prime(limit):
    sum_of_prims=[]
    sum=0
    prime_list = get_primes(limit)
    for prime in prime_list:
        sum+=prime
        if sum in prime_list and sum<limit:
            sum_of_prims.append(sum)
            
    return {"largest_prime":sum_of_prims[-1]}  
```

```python
find_largest_prime(10000)
```
